```
using System;
using System.Collections.Generic;
using System.Linq;
using System.Text;
using System.Threading.Tasks;

namespace ConsoleApp1
{
    class Matrix
    {
        int n;
        int[,] mas;
        public Matrix(int n)
        {
            this.n = n;
            mas = new int[n, n];
            Random r = new Random();
            for(int i = 0; i < n;i++)
            {
                for (int j = 0; j < n; j++)
                {
                    mas[i, j] = r.Next(2);
                }
            }
        }
        public void Fill(int x, int y ,int c,int f)
        {
            if(x>=0 && x<n && y >=0 && y < n && mas[x, y] == f)
            {
                mas[x, y] = c;
                Fill(x-1, y, c, f);
                Fill(x, y-1, c, f);
                Fill(x+1, y, c, f);
                Fill(x, y+1, c, f);

            }
        }
        public void Print()
        {
            string st = "";
            for (int i = 0; i < n; i++)
            {
                for (int j = 0; j < n; j++)
                {
                    st+=mas[i, j].ToString();
                }
                st += "\n";
            }
            Console.WriteLine(st);
        }
        public Dictionary<int, int> Count()
        {
            Dictionary<int, int> dict = new Dictionary<int, int>();
            for (int i = 0; i < n; i++)
            {
                for (int j = 0; j < n; j++)
                {
                    if (dict.ContainsKey(mas[i, j]))
                    {
                        dict[mas[i, j]] += 1;
                    }
                    else
                    {
                        dict.Add(mas[i, j], 1);
                    }
                }
            }
            return dict;
        }
    }
    class Program
    {
        static void Main(string[] args)
        {
            Matrix m = new Matrix(10);
            //m.Print();
            m.Fill(0, 1, 5, 0);
            //m.Print();
            var dict = m.Count();
            foreach (int c in dict.Keys)
            {
                Console.WriteLine("{0} - {1}",c, dict[c]);
            }
            Console.ReadKey();

        }
    }
}
```
